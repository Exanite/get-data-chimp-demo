```python
import pandas as pd
import seaborn as sns
import random
```

```python
df = pd.read_csv('iris.csv')
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='variety', data=df)
```

```python
sns.scatterplot(x='petal_length', y='petal_width', hue='variety', data=df)
```

```python
s = pd.Series(1 if random.random() > .5 else 10 for n in range(150))
```

```python
df.groupby('variety').min()
```

```python
df.loc[df['variety'] == 'Versicolor', 'petal_length'] *= s[df['variety'] == 'Versicolor']
df.loc[df['variety'] == 'Versicolor', 'petal_width'] *= s[df['variety'] == 'Versicolor']
df.loc[df['variety'] == 'Versicolor', 'collector'] = 'Eric'
df['collector'].fillna('Sam', inplace=True)
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='variety', data=df)
```

```python
sns.scatterplot(x='petal_length', y='petal_width', hue='variety', data=df)
```

```python
df.to_csv('borked_iris.csv')
df.to_csv('../iris.csv')
```
